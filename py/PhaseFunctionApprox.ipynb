```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
# core scatterings
mu = np.linspace(-1,1,1000)  # values of mu = cos(theta), between -1 and 1
wc = 0.5 + 11./24*mu + 1./24*mu**3 # cumulative probability (P(<mu)).
```

```python
# plot cumultive proba vs. mu
plt.figure()
plt.plot(mu,wc)
plt.xlim(-1,1)
plt.ylim(0,1)
plt.xlabel(r'$\mu=cos(\theta)$')
plt.ylabel(r'$P(<\mu)$')
```

```python
plt.figure()
for n in [3,5]:
    p = np.polyfit(wc,mu,n)
    relerr = np.abs( (np.polyval(p,wc)-mu) /mu)
    plt.plot(wc,relerr*100.,label='n = %i'%n)
    print 'max relative error : %f per cent'%(relerr.max()*100.)
    print ["%f x^%i"%(pp,n-i) for i,pp in enumerate(p)]
plt.yscale('log')
plt.legend()
plt.xlabel(r'$P(<\mu)$')
plt.ylabel('rel. error [%]')
```

```python
# chosen approx (better than ~0.5% everywhere)
mufit = -0.703204 * wc**3 + 1.054807 * wc**2 + 1.643182*wc - 0.997392
plt.plot(mu,wc,color='blue',lw=3,alpha=0.7)
plt.plot(mufit,wc,color='red')
```
